Tạo mảng 20 số nguyên ngẫu nhiên kích thước 3 x 3 có giá trị trong khoảng [0, 20]. Viết
chương trình kiểm tra tất cả các phần tử trong mảng khác 0 (tìm hiểu hàm all của numpy)

```python
import numpy as np
b1 = np.random.randint(20, size = (3,3))
print (b1)

if np.all(b1) != 0:
  print ('Khong co phan tu 0')
else:
  print ('Co phan tu 0')
```
